# 1. Посмотреть документацию к API GitHub, разобраться как вывести список репозиториев для конкретного пользователя, сохранить JSON-вывод в файле *.json.

```python
import requests
import json
```

```python
response = requests.get(url="https://api.github.com/users/gusujio/repos") #для получения списка репозиториев достаточно знать логин
```

```python
response.json()
```

```python
with open('data_1.json', 'w') as outfile:
    json.dump(response.json(), outfile)
```

```python
for i in range(len(response.json())):
    print(response.json()[i]['id'], response.json()[i]['name'])
```
